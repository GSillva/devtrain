```python
import pandas as pd
import math

# Load the recently created spreadsheet (copy)
df = pd.read_excel('/content/Engenharia de Software - Desafio [Gabriel Silva].xlsx')

# Define total classes per semester
total_classes_per_semester = 60

# Define a function to calculate the situation based on average and number of absences
def calculate_situation(row):
    avg = (row['P1'] + row['P2'] + row['P3']) / 3
    absences = row['Faltas']

    if absences > total_classes_per_semester * 0.25:
        row['Situation'] = 'Reprovado por Falta'
        row['Nota para Aprovação Final'] = 0
    elif avg < 5:
        row['Situation'] = 'Reprovado por Nota'
        row['Nota para Aprovação Final'] = 0
    elif 5 <= avg < 7:
        row['Situation'] = 'Exame Final'
        naf = math.ceil((10 - avg) * 2)
        if naf % 2 != 0:
            naf += 1  # Round up to the next even number
        row['Nota para Aprovação Final'] = naf
    else:
        row['Situation'] = 'Aprovado'
        row['Nota para Aprovação Final'] = 0

    return row

# Apply the function to each row in the DataFrame
df = df.apply(calculate_situation, axis=1)

# Save the updated DataFrame to a new spreadsheet
df.to_excel('/content/updated_spreadsheet.xlsx', index=False)

df1 = pd.read_excel('/content/updated_spreadsheet.xlsx')
df1
```
